```python
import cv2
import numpy as np
from ultralytics import YOLO
from collections import defaultdict
from sklearn.cluster import DBSCAN
import math

model = YOLO('/kaggle/input/trained-yolo/working/runs1/detect/yolov8n_custom/weights/best.pt')
video_path = "/kaggle/input/prediction-video/gender.mp4"
cap = cv2.VideoCapture(video_path)

distance_threshold = 50
tracker = defaultdict(lambda: {"centroid": None, "bbox": None, "class": None, "id": None, "frames_missing": 0})
object_id = 0

def calculate_centroid(x1, y1, x2, y2):
    return ((x1 + x2) // 2, (y1 + y2) // 2)

def calculate_distance(point1, point2):
    return math.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)

output_video_path = 'output_video.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    current_frame_objects = []

    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            class_id = int(box.cls[0].item())
            centroid = calculate_centroid(x1, y1, x2, y2)
            current_frame_objects.append((centroid, (x1, y1, x2, y2), class_id))

    men_coordinates = []
    women_coordinates = []

    for centroid, bbox, class_id in current_frame_objects:
        if class_id == 1:  # Man
            men_coordinates.append(centroid)
        elif class_id == 0:  # Woman
            women_coordinates.append(centroid)

    # DBSCAN Clustering for the Current Frame
    coordinates = np.array(men_coordinates + women_coordinates)
    if len(coordinates) > 0:
        dbscan = DBSCAN(eps=50, min_samples=2).fit(coordinates)
        labels = dbscan.labels_

        for cluster_id in set(labels):
            if cluster_id == -1:
                continue
            cluster_indices = np.where(labels == cluster_id)[0]
            cluster_points = coordinates[cluster_indices]

            men_in_cluster = sum([1 for i in cluster_indices if i < len(men_coordinates)])
            women_in_cluster = sum([1 for i in cluster_indices if i >= len(men_coordinates)])

            # Alert Conditions for Current Frame
            if women_in_cluster > 0:
                ratio = women_in_cluster / (men_in_cluster + 1e-5)
                if ratio < 0.25 and women_in_cluster < 5:
                    print("Alert: Low women-to-men ratio and less than 5 women detected in this cluster.")

    # Display counts and annotations on the frame
    cv2.putText(frame, f'Men: {len(men_coordinates)}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
    cv2.putText(frame, f'Women: {len(women_coordinates)}', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    out.write(frame)

cap.release()
out.release()

print("Processing complete. Check the generated video.")
```
